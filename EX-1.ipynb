```python
# Plot the scree plot and cumulative explained variance
import numpy as np
import matplotlib.pyplot as plt

variance_ratios = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(variance_ratios)
components = np.arange(1, len(variance_ratios) + 1)

plt.figure(figsize=(10, 6))
plt.bar(components, variance_ratios, alpha=0.7, color='blue', label='Individual explained variance')
plt.plot(components, cumulative_variance, 'o-', color='red', label='Cumulative explained variance')
plt.title('Scree Plot with Cumulative Explained Variance')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(components)
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Normaly, it is easier to see the corr values in the heatmaps
##plot the heatmap

import seaborn as sns

plt.figure(figsize=(18, 18))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
### Finding out Strongly Correlated features:

# Create an empty DataFrame to store the upper triangle matrix
upper_tri = pd.DataFrame(np.nan, index=corr_matrix.index, columns=corr_matrix.columns)

for i in range(corr_matrix.shape[0]):
    for j in range(corr_matrix.shape[1]):
        if i < j:
            upper_tri.iloc[i,j]=corr_matrix.iloc[i,j]
```

```python
## upper trianle of correlation matrix

upper_tri
```

```python
#### To determine strongly correlated features

ths_value=0.8
to_drop=[]

for column in upper_tri.columns:
    if any(upper_tri[column] > ths_value):
        to_drop.append(column)

print(to_drop)
```

```python
# Drop features
df_reduced = df.drop(to_drop, axis=1)
```

### Here I will do stratified Sampling for K-fold Cross validation
### To maintain the same ratio of classes in both train and test sets

```python
from collections import Counter

def ratio_check(list):
    number=Counter(list)
    
    ratio_two_classes=number[0]/number[1]
    return ratio_two_classes
```

```python
### set up stratified K-fold

from sklearn.model_selection import StratifiedKFold 

#5fold CV
n_splits=5
skf=StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)


### Now apply this 5 fold CV to split train and test dataset

for train_index, test_index in skf.split(df_, target):
    X_train, x_test=df_.iloc[train_index], df_.iloc[test_index]
    Y_train, y_test=target[train_index], target[test_index]
    
    print('Ratio in Training Set', ratio_check(Y_train), 'Ratio in Test Set', ratio_check(y_test))
    #print(train_index, test_index)
```

### If there is a imbalance in classes in the dataset, then we need to do resampling.

#### Resampling are two kinds: (1) Oversampling and (2) Undersampling.

### I used Random Resampling for our Datasets

#### Random Oversampling: Randomly duplicate examples in the minority class.
#### Random Undersampling: Randomly delete examples in the majority class.


#### The model is evaluated using repeated 5-fold cross-validation, and the oversampling is performed on the training dataset within each fold separately, ensuring that there is no data leakage as might occur if the oversampling was performed prior to the cross-validation.

```python

#Imp Libraries
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, f1_score, recall_score, roc_curve, auc
import pandas as pd
import matplotlib.pyplot as plt

#Load of cancer dataset
cancer=load_breast_cancer()
df_cancer=pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
target=cancer.target


#Dimentionality Reduction using PCA with variance 80%
#Applying StandradScaler normalization before applying PCA
df_cancer=StandardScaler().fit_transform(df_cancer)
data_pca=PCA(0.8).fit_transform(df_cancer)

#Converting into a dataframe
df_pca=pd.DataFrame(data=data_pca)

#Applying stratified kfold (5 fold) Crossvalidation
skf=StratifiedKFold(n_splits=5,shuffle=True, random_state=42)

#intitialization of lists and inetgeer
fold=0
all_y_gt = []
all_y_pred = []
all_y_prob = []

for ind_train, ind_test in skf.split(df_pca,target):
    X_train, x_test = df_pca.iloc[ind_train], df_pca.iloc[ind_test]
    Y_train, y_test = target[ind_train], target[ind_test]
    
    #oversampling the minority classes in TRAINING SET using imblearn lib
    ovs=SMOTE(sampling_strategy='minority')
    X_ovs, Y_ovs = ovs.fit_resample(X_train, Y_train)
    
    #Random Forest Classifier
    model=RandomForestClassifier()
    model.fit(X_ovs, Y_ovs)
    
    #predicting on test set
    Y_pred=model.predict(x_test)
    
    #Calculating probabilities for AUCROC
    Y_prob=model.predict_proba(x_test)[:,1]
    
    #Performance report
    #report=classification_report(y_test, Y_pred)
    pre=precision_score(y_test, Y_pred)
    recall=recall_score(y_test, Y_pred)
    f1=f1_score(y_test, Y_pred)
    
    fpr, tpr, _= roc_curve(y_test, Y_prob)
    print('######')
    fold=fold+1
    print('fold =', fold)
    print()
    print('precision_score =', pre, 'recall_score =', recall, 'f1 score =', f1  )
    print()
    #Area under the cureve
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    #plt.plot([0, 1], [0, 1], color='grey', lw=2, linestyle='--')  # Diagonal line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
    print()
    
    
    #Saving all predictions from each fold
    all_y_gt.extend(y_test)
    all_y_pred.extend(Y_pred)
    all_y_prob.extend(Y_prob)
    
```

```python
#final classification metrics for the Entire dataset
import numpy as np

all_y_gt=np.array(all_y_gt)
all_y_pred=np.array(all_y_pred)
all_y_prob= np.array(all_y_prob)

pre=precision_score(all_y_gt, all_y_pred)
recall=recall_score(all_y_gt, all_y_pred)
f1=f1_score(all_y_gt, all_y_pred)
    
fpr, tpr, _= roc_curve(all_y_gt, all_y_prob)

print('######')
print()
print('precision_score =', pre, 'recall_score =', recall, 'f1 score =', f1  )
print()
    
#Area under the cureve
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
#plt.plot([0, 1], [0, 1], color='grey', lw=2, linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
print()


```
